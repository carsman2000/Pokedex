```python
import sqlite3
import numpy as np
import pandas as pd
```

```python
info = pd.read_csv('Book1.csv')
con = sqlite3.connect("C:/Users/Cameron/Documents/Java Projects/Pokedex/pokeDatabase.db")
cur = con.cursor()
for i in range(151):
    type1 = info['type1'][i]
    type2 = info['type2'][i]
    ability1 = info['ability1ID'][i]
    ability2 = info['ability2ID'][i]
    hiddenAbility = info['hiddenAbilityID'][i]
    
    type2ID = -1
    ability2ID = -1
    hiddenAbilityID = -1
    
    type1ID = con.execute("SELECT typeID FROM Types WHERE name = \"" + type1 + "\"").fetchone()[0]
    if(isinstance(type2, str)):
        type2ID = con.execute("SELECT typeID FROM Types WHERE name = \"" + type2 + "\"").fetchone()[0]
    
    ability1ID = con.execute("SELECT abilityID FROM Abilities WHERE name = \"" + ability1 + "\"").fetchone()[0]
    if(isinstance(ability2, str)):
        ability2ID = con.execute("SELECT abilityID FROM Abilities WHERE name = \"" + ability2 + "\"").fetchone()[0]
    if(isinstance(hiddenAbility, str)):
        hiddenAbilityID = con.execute("SELECT abilityID FROM Abilities WHERE name = \"" + hiddenAbility + "\"").fetchone()[0]
    
    command = "INSERT INTO Pokemon (dexNum, name, species, type1, type2, ability1ID, ability2ID, hiddenAbilityID, height, weight) VALUES ("
    command += str(info['dexNum'][i]) + ", "
    command += "\"" + info['name'][i] + "\", "
    command += "\"" + info['species'][i] + "\", "
    command += str(type1ID) + ", "
    
    if(type2ID != -1):
        command += str(type2ID) + ", "
    else:
        command += "NULL, "
        
    command += str(ability1ID) + ", "
    
    if(ability2ID != -1):
        command += str(ability2ID) + ", "
    else:
        command += "NULL, "
        
    if(hiddenAbilityID != -1):
        command += str(hiddenAbilityID) + ", "
    else:
        command += "NULL, "
        
    command += str(info['height'][i]) + ", "
    command += str(info['weight'][i]) + ");"
    
    print(command)
    con.execute(command)
    
con.commit()
con.close()
```
